```python
#Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.
```

```python
import numpy as np
import seaborn as sns

tips=sns.load_dataset("tips")
tips
```

```python
#1 PLOT 
sns.scatterplot(x=tips.total_bill,y=tips.tip)
```

```python
# second plot
sns.displot("total_bill")
```

```python
# third plot
sns.relplot(x=tips.total_bill,y=tips.tip,data=tips,style="size")
```

```python
# fourth plot
sns.catplot(x="sex",y="tip",data=tips)
```

```python
# fifth plot
sns.pairplot(tips)
```

```python
#Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.
```

```python
import seaborn as sns
import numpy as np
fmri=sns.load_dataset("fmri")
sns.lineplot(x = "timepoint", y = "signal", data=fmri, hue="event",size="region")
```

```python
#Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.
```

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

titanic=sns.load_dataset("titanic")
#plot ist box plot
plt.figure(figsize=(10,6))
sns.boxplot(x="pclass" , y="age",data=titanic)
plt.title("box plot age by pclass")
plt.xlabel("pclass")
plt.ylabel("age")
plt.show()
```

```python
#plot second box plot
plt.figure(figsize=(10,6))
sns.boxplot(x="pclass" , y="fare",data=titanic)
plt.title("box plot fare by pclass")
plt.xlabel("pclass")
plt.ylabel("fare")
plt.show()
```

```python
#Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
#parameter for the 'cut' column of the diamonds dataset.
```

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

diamonds=sns.load_dataset("diamonds")

sns.histplot(x="price",hue="cut",data=diamonds)
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Histogram of Diamond Prices by Cut')
plt.show()
```

```python
#Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
#of the iris dataset.
```

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

iris=sns.load_dataset("iris")
sns.pairplot(data=iris,hue="species")
plt.title("this is pairplot of iris")
plt.show()
```

```python
#Que 6: Use the "flights" dataset from seaborn to plot a heatmap.
```

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

flights=sns.load_dataset("flights")
flights_pivot=flights.pivot_table(index="month", columns="year", values="passengers")
plt.figure(figsize=(10, 6))
sns.heatmap(flights_pivot, cmap="YlGnBu", annot=True, fmt="d")
plt.title("Flights Heatmap")
plt.show()
```
