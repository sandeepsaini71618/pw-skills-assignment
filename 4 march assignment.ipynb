```python
#Q1. Load the "titanic" dataset using the load_dataset function of seaborn. Use Plotly express to plot a scatter plot for age and fare columns in the titanic dataset.
```

```python
import seaborn as sns
titanic=sns.load_dataset("titanic")
titanic
```

```python
import plotly.graph_objects as go
fig=go.Figure()
fig.add_trace(go.Scatter(x=titanic.age,y=titanic.fare,mode="markers"))
fig.show()
```

```python
#Q3. Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill" column in
#the tips dataset. Also, use the "smoker" column with the pattern_shape parameter and the "day"
#column with the color parameter.
```

```python
import plotly.express as px

#load the data sheet
tips_data=px.data.tips()

#create a histogram
histogram= px.histogram(tips_data,
    x="sex",
    y="total_bill",
    color="day",
    pattern_shape="smoker",
    title="Histogram of Total Bill by Sex with Smoker and Day Information")

# Show the plot
histogram.show()
```

```python
#Q4. Using the iris dataset in the Plotly library, Plot a scatter matrix plot, using the "species" column for the color parameter.

#Note: Use "sepal_length", "sepal_width", "petal_length", "petal_width" columns only with the dimensions parameter.
```

```python
#Q5. What is Distplot? Using Plotly express, plot a distplot.
```

A distplot, short for "distribution plot," is a data visualization that combines both a histogram and a kernel density estimate (KDE) plot. It is used to visualize the distribution of a univariate dataset, which means it's suitable for exploring the distribution of a single variable.

A distplot typically shows the following information:

Histogram: A bar chart that represents the frequency or count of data points within specific bins or intervals along the variable's range.

Kernel Density Estimate (KDE): A smoothed line that provides an estimate of the probability density function of the data. It offers a continuous representation of the data distribution.

```python
import plotly.express as px

# Create a sample dataset
data = [1, 1, 1, 2, 2, 2, 3, 3, 4, 4, 4, 5, 5, 6]

# Create a distplot
distplot = px.histogram(data, marginal="rug", title="Distplot Example")

# Show the plot
distplot.show()
```
