```python
#Q2. What are the types of Probability distribution?
```

 there are several types of probability distributions. These distributions describe how the probabilities of different outcomes are spread or distributed. Here are some common types of probability distributions with explanations and examples:

Uniform Distribution:

A uniform distribution occurs when all outcomes are equally likely.
Example: Rolling a fair six-sided die. Each face has a probability of 1/6.

Bernoulli Distribution:

A Bernoulli distribution models a binary outcome, such as success or failure.
Example: Flipping a coin, where success might be getting heads (p) and failure getting tails (1-p).

Binomial Distribution:

A binomial distribution models the number of successes in a fixed number of independent Bernoulli trials.
Example: The number of heads obtained when flipping a coin 10 times with a probability of 0.5 for heads.

Poisson Distribution:

A Poisson distribution models the number of events occurring in a fixed interval of time or space.
Example: The number of emails received in an hour, where the average rate is 5 emails per hour.

Normal Distribution (Gaussian Distribution):

The normal distribution is bell-shaped and is characterized by its mean and standard deviation.
Example: Heights of adult humans, IQ scores, and many naturally occurring phenomena.

```python
#Q4. What are the properties of Binomial distribution? Give two examples of events where binomial distribution can be applied.
```

The binomial distribution is a probability distribution that describes the number of successes in a fixed number of independent Bernoulli trials, where each trial has the same probability of success, denoted by "p." It is characterized by the following properties:

Two Possible Outcomes: In a binomial experiment, there are only two possible outcomes for each trial: success or failure.

Fixed Number of Trials (n): The number of trials (n) is predetermined and remains constant throughout the experiment.

Independence: Each trial is independent of the others, meaning that the outcome of one trial does not affect the outcome of any other trial.

Constant Probability (p): The probability of success (p) is the same for each trial.

Discrete: The binomial distribution is a discrete probability distribution, meaning that it deals with countable, distinct outcomes (0, 1, 2, ...).

Probability Mass Function (PMF): The probability mass function of the binomial distribution is given by the formula: P(X = k) = C(n, k) * p^k * (1-p)^(n-k), where P(X = k) is the probability of getting exactly k successes, C(n, k) is the binomial coefficient (combinations), p is the probability of success, and (1-p) is the probability of failure.

Two examples of events where the binomial distribution can be applied are:

Coin Flips: When flipping a fair coin (where the probability of getting heads or tails is 0.5) multiple times, we can use the binomial distribution to calculate the probability of getting a specific number of heads (successes) in a fixed number of flips.

Quality Control: In manufacturing, the binomial distribution can be applied to assess the probability of a certain number of defective items in a sample from a production batch. For example, if you randomly select 10 items from a production line where each item has a 5% chance of being defective, you can use the binomial distribution to determine the likelihood of finding a particular number of defective items in the sample.

```python
#Q6. Write a Python function to calculate the cumulative distribution function of a Poisson distribution with given mean at a given point.
```

```python
import math 
def poisson_cdf(x,mean):
    cdf=0.0
    for k in range(x+1):
        cdf+=(math.exp(-mean)*mean**k)/math.factorial(k)
    return cdf

mean=5
x=3
result=poisson_cdf(x,mean)
print(f"The poisson cdf at x={x} with mean={mean} is {result } ")
```

```python
#Q7. How Binomial distribution different from Poisson distribution?
```

The Binomial distribution and the Poisson distribution are both probability distributions used in statistics and probability theory, but they are distinct in several ways. Let me explain the key differences between them:

Nature of Events:

Binomial Distribution: It is used to model discrete events that have exactly two possible outcomes, often referred to as "success" and "failure." Each trial in a binomial distribution is independent and has the same probability of success.
Poisson Distribution: It is used to model the number of events that occur in a fixed interval of time or space, where these events happen randomly and independently. It does not have a fixed number of trials like the binomial distribution.

Number of Trials:

Binomial Distribution: It requires a fixed number of trials (n) to be specified in advance.
Poisson Distribution: It does not require a fixed number of trials; instead, it focuses on events occurring in a fixed interval.

Parameters:

Binomial Distribution: It depends on two parameters: the probability of success (p) and the number of trials (n).
Poisson Distribution: It depends on a single parameter, which is the average rate of event occurrences (λ) in the given interval.

Probability Mass Function (PMF):

Binomial Distribution: The PMF of the binomial distribution gives the probability of getting exactly k successes in n trials.
Poisson Distribution: The PMF of the Poisson distribution gives the probability of observing k events in the fixed interval.

Use Cases:

Binomial Distribution: It is commonly used when dealing with situations like coin flips, where there are two possible outcomes (heads or tails), and we want to calculate the probability of getting a specific number of heads in a fixed number of flips.
Poisson Distribution: It is often used to model rare events or the number of occurrences of events in a fixed time or space, such as the number of customer arrivals at a store in an hour.

```python
#Q9. How mean and variance are related in Binomial distribution and Poisson distribution?
```

In statistics and probability theory, mean and variance are two fundamental measures used to describe the properties of a probability distribution. Let's explore how mean and variance are related in the context of the Binomial distribution and the Poisson distribution.

Binomial Distribution:
The Binomial distribution models the number of successes (usually denoted as "k") in a fixed number of independent and identical Bernoulli trials (experiments with two possible outcomes: success and failure). The mean and variance of a Binomial distribution are related as follows:

Mean (μ): The mean of a Binomial distribution is given by the formula: μ = n * p, where "n" is the number of trials and "p" is the probability of success in each trial.

Variance (σ^2): The variance of a Binomial distribution is given by the formula: σ^2 = n * p * (1 - p), where "n" is the number of trials, and "p" is the probability of success in each trial.

In summary, for a Binomial distribution, the variance is directly related to both the number of trials (n) and the probability of success (p). As the number of trials or the probability of success changes, the variance will also change accordingly.

Poisson Distribution:
The Poisson distribution models the number of events (usually denoted as "k") that occur in a fixed interval of time or space when events are rare and random. The mean and variance of a Poisson distribution are related as follows:

Mean (μ): The mean of a Poisson distribution is equal to the rate of event occurrences in the given interval, often denoted as "λ" (lambda). Therefore, μ = λ.

Variance (σ^2): The variance of a Poisson distribution is also equal to λ. Hence, σ^2 = λ.

In summary, for a Poisson distribution, the mean and variance are both equal to the rate parameter λ. This means that in a Poisson distribution, the variance is directly related to the mean, and they have the same value.

In a normal distribution, also known as a Gaussian distribution, the least frequent data points appear in the tails of the distribution, farthest away from the mean position. These tails are the regions where the data values are significantly higher or lower than the mean. In a standard normal distribution (with a mean of 0 and a standard deviation of 1), about 68% of the data falls within one standard deviation of the mean, while the remaining 32% is distributed in the tails, with approximately 16% in the lower tail (below -1 standard deviation) and 16% in the upper tail (above +1 standard deviation).
