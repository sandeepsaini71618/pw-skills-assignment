```python
#1 What is Flask Framework? What are the advantages of Flask Framework?
```

Flask is a micro web framework written in Python. It is designed to be lightweight, simple, and easy to use, allowing developers to quickly build web applications and APIs. Flask follows the WSGI (Web Server Gateway Interface) standard, which makes it compatible with various web servers and platforms.

Advantages of Flask Framework:

Simplicity: Flask provides a straightforward and minimalist approach to web development. Its core is small and easy to understand, making it ideal for beginners and small-scale projects.

Flexibility: Flask doesn't enforce any specific way to structure your application, giving developers the freedom to choose how to organize code and project components. This flexibility makes it adaptable to various development styles and project requirements.

Extensibility: Flask offers a wide range of extensions, also known as Flask extensions, that provide additional functionality to the framework. These extensions make it easy to add features like database integration, authentication, and caching, without having to reinvent the wheel.

Minimal Boilerplate Code: The simplicity of Flask means you write less boilerplate code to get your application up and running. This results in more focused and concise code, making development quicker and easier.

Well-Documented: Flask has excellent documentation that covers various aspects of the framework. This makes it easier for developers to learn and understand how to use Flask effectively.

Large Community: Flask has a vibrant and active community of developers. This means you can find plenty of resources, tutorials, and third-party packages to help with your project. Additionally, the community actively supports and maintains the framework, ensuring continuous improvement.

Testing Support: Flask provides built-in testing support, allowing developers to write unit tests and integration tests for their applications. This ensures better code quality and maintainability.

Performance: Being a lightweight framework, Flask has lower overhead, resulting in better performance compared to more heavyweight frameworks. It's an excellent choice for building high-performance applications and APIs.

```python
#3  What is App routing in Flask? Why do we use app routes?

```

Flask, app routing refers to the process of defining URL patterns (routes) that map to specific functions in your application. When a user makes a request to a specific URL, Flask uses the defined routes to determine which function should handle that request. This allows you to create different endpoints for different parts of your application, making it easier to organize and manage your code.

We use app routes in Flask for the following reasons:



Handling Different URLs: With app routes, you can create different endpoints for various parts of your web application. For example, you might have routes for the home page, user login, user registration, and other functionalities.

Request Methods: App routes allow you to specify which HTTP methods (e.g., GET, POST, PUT, DELETE) a particular route should respond to. This helps in creating RESTful APIs and handling different types of user interactions.



View Functions: Each app route is associated with a view function that handles the logic for that specific URL. When a request is made to a route, Flask calls the corresponding view function, and the response from that function is sent back to the client.



```python
from flask import Flask

app=Flask(__name__)

#for home page
@app.route("/")
def home_page():
    return "Thats is my home page"

#define for about page
@app.route("/about-us")
def about():
    return "this is about us page"

#accepts a dynamic parameter
@app.route("/use/<username>")
def userprofile(username):
    return f"hello , {username}"

#handles form submissions
@app.route("/submit", methods=["POST"])
def form_submission():
    return "form submitted successfully"

if __name__=="__main__":
    app.run(host="0.0.0.0")
```

```python
#5 What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
#url_for() function.
```

In Flask, the function used for URL building is url_for(). This function is helpful when you want to generate URLs dynamically for different views in your web application. It allows you to avoid hardcoding URLs in your templates or code, making your application more maintainable and flexible.

Here's a Python code demonstrating the working of the url_for() function:

```python
from flask import Flask,url_for

app=Flask(__name__)

@app.route("/")
def home_page():
    return "This is the home page"

@app.route("/about_us")
def about_us():
    return "This is the about us page"


@app.route("/contact_us")
def contact_us():
    return "This is the contact us page"

if __name__=="__main__":
    with app.test_request_context():
        
        home_url=url_for("home_page")
        about_url=url_for("about_us")
        contact_url=url_for("contact_us")
        
        print(f'Home URL: {home_url}')
        print(f'About URL: {about_url}')
        print(f'Contact URL: {contact_url}')
```
